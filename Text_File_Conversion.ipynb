```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dict={'C':[12,20,34,45,55],
      'c++':[15,25,36,50,60],
      'java':[18,30,40,55,70],
      'r':[45,50,55,60,75],
      'python':[55,60,70,80,100]}

df=pd.DataFrame(dict, index=['200Mb','400Mb','600Mb','800Mb','1000Mb'])
print(df)

plt.plot(df, marker='o')
plt.grid(True)
plt.title("text conversion")
plt.xlabel('file sizes')
plt.ylabel('time taken')
plt.legend(df.columns, loc='lower right')
plt.show()
```
